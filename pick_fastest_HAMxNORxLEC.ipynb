```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import fastf1
```

```python
MI = fastf1.get_session(2024, 6, 'R')
MI.name
```

```python
MI.event
```

```python
MI.load()
MI.results
```

```python
import fastf1.plotting
fastf1.plotting.setup_mpl(misc_mpl_mods=False)
```

```python
ver_lap = MI.laps.pick_driver('VER').pick_fastest()
nor_lap = MI.laps.pick_driver('NOR').pick_fastest()
```

```python
ver_tel = ver_lap.get_car_data().add_distance()
nor_tel = nor_lap.get_car_data().add_distance()
```

```python
rbr_color = fastf1.plotting.team_color('RBR')
mcl_color = fastf1.plotting.team_color('MCL')

fig, ax = plt.subplots()
ax.plot(ver_tel['Distance'], ver_tel['Speed'], color=rbr_color, label='VER')
ax.plot(nor_tel['Distance'], nor_tel['Speed'], color=mcl_color, label='NOR')

ax.set_xlabel('Distância em m')
ax.set_ylabel('Velocidade km/h')

ax.legend()
plt.suptitle(f"Comparação de Volta mais rápida \n "
             f"{MI.event['EventName']} {MI.event.year}")
plt.show()
```

```python
ham_lap = MI.laps.pick_driver('HAM').pick_fastest()
```

```python
ham_tel = ham_lap.get_car_data().add_distance()
```

```python
mer_color = fastf1.plotting.team_color('MER')
mcl_color = fastf1.plotting.team_color('MCL')

fig, ax = plt.subplots()
ax.plot(ham_tel['Distance'], ham_tel['Speed'], color=mer_color, label='HAM')
ax.plot(nor_tel['Distance'], nor_tel['Speed'], color=mcl_color, label='NOR')

ax.set_xlabel('Distância em m')
ax.set_ylabel('Velocidade km/h')

ax.legend()
plt.suptitle(f"Comparação de Volta mais rápida \n "
             f"{MI.event['EventName']} {MI.event.year}")
plt.show()
```
